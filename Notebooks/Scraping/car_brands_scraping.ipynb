```python
# Get the current date
current_date = datetime.now().strftime("%d_%m_%Y")  
cd = datetime.now().strftime("%m_%d_%Y") 
# Capture the start time for postings scraping
start_time_scraping_CG_posts = datetime.now()

# Use EdgeOptions instead of the general Options
# options = Options()
options = webdriver.EdgeOptions()
driver = webdriver.Edge(executable_path='C:/Windows/System32/msedgedriver.exe',options=options)
# Correctly set the binary location for Edge
# options.binary_location = r'C:\Windows\System32\msedge.exe'  # Update with your actual Edge path

options.add_argument("--window-size=1920x1080")
options.add_argument("--headless")
options.add_argument("--verbose")

# Explicitly set the Edge binary location and path to msedgedriver
# service = Service(executable_path=r'C:\Windows\System32\msedgedriver.exe')  # Update with your actual msedgedriver path
# browser = webdriver.Edge(service=service, options=options)
driver.set_page_load_timeout(30)

data = []


URL = f'https://www.carlogos.org/car-brands-a-z/'
driver.get(URL)

all_content = driver.find_element(By.CSS_SELECTOR, 'div.a-z')  # Use find_element to find the container div

dd_elements = all_content.find_elements(By.CSS_SELECTOR, 'dd')  # Get all dd elements

for dd_element in dd_elements:
    try:
        link_element = dd_element.find_element(By.TAG_NAME, 'a')
        brand_name = link_element.text
        print(brand_name)
        brand_link = link_element.get_attribute('href')
        print(brand_link)
    except NoSuchElementException:
        print(f"No link found in the element, skipping: {dd_element.text}")
    else:
        data.append({
            'Brand_name': brand_name,
            'Brand_Content_URL': brand_link
        })

driver.quit()

car_brands_df = pd.DataFrame(data)
print(car_brands_df)
```

```python

# ... (Your other imports and setup remain the same)

def logo_link(url, brand_name):
    """Downloads and saves the logo image from the given URL.

    Args:
        url (str): The URL of the webpage containing the logo.
        brand_name (str): The name of the brand.

    Returns:
        str or None: The filename of the saved image, or None if an error occurred.
    """
    # Use EdgeOptions instead of the general Options
    # options = Options()
    options = webdriver.EdgeOptions()
    driver = webdriver.Edge(executable_path='C:/Windows/System32/msedgedriver.exe',options=options)
    options.add_argument("--window-size=1920x1080")
    options.add_argument("--headless")
    options.add_argument("--verbose")
    driver.set_page_load_timeout(30)

    try:
            driver.get(url)

            brand_content = driver.find_element(By.CSS_SELECTOR, "div.content")
            logo_link_element = brand_content.find_element(By.TAG_NAME, "a")
            return logo_link_element.get_attribute('href')
    except Exception as e:
        print(f"Error processing {brand_name}: {e}")
        return None
    finally:
        driver.quit()  # Close the driver after processing each URL

# ... (Your code to scrape brand names and URLs into car_brands_df) ...
# Assuming your DataFrame has a column named 'Brand_Content_URL'

```

```python
car_brands_df['Brand_logo_URL'] = car_brands_df['Brand_Content_URL'].apply(
    lambda url: logo_link(url, car_brands_df.loc[car_brands_df['Brand_Content_URL'] == url, 'Brand_name'].iloc[0]) 
)
```
