```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET

url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetAllManufacturers'
response = requests.get(url)

# Parse the XML response
root = ET.fromstring(response.content)

# Extract relevant data, handling missing values
data = []
for manufacturer in root.findall('.//Manufacturers'):
    data.append({
        'Mfr_CommonName': manufacturer.find('Mfr_CommonName').text if manufacturer.find('Mfr_CommonName') is not None else None,
        'Mfr_Name': manufacturer.find('Mfr_Name').text if manufacturer.find('Mfr_Name') is not None else None
    })

# Create DataFrame
df = pd.DataFrame(data)

print(df)
```
