```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
# Convert all text to lowercase
complaints_df['Complaint'] = complaints_df['Complaint'].str.lower()

# Tokenize the text
#complaints_df['tokens'] = complaints_df['Complaint'].apply(word_tokenize())
complaints_df['tokens'] = complaints_df['Complaint'].apply(lambda x: [word for word in nltk.word_tokenize(x) if len(word)>3])

# Remove stopwords
stop_words = set(stopwords.words('english'))
complaints_df['tokens'] = complaints_df['tokens'].apply(lambda x: [word for word in x if word not in stop_words])

# Join the tokens back into text
complaints_df['processed_text'] = complaints_df['tokens'].apply(lambda x: ' '.join(x))
```
