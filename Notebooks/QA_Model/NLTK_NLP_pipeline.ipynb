#### <u>Tokenization:</u> Split up a text by word or by sentence.
##### By word: Tokenizing a text by word allow us to understand important/strong words in the sentence alongside filtering out on the stop words.
##### By sentence: Tokenizing a text by sentence would bring context into understanding a word and how group of words relate to one another.


```python
import nltk
nltk.download()

from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# An example string
example_string = """ Muad'Dib learned rapidly because his first training was in how to learn. And the first lesson of all was the basic trust that he could learn. It's shocking to find how many people do not believe they can learn, and how many more believe learning to be difficult."""
```

```python
sent_tokenize(example_string)
# Applying sentence tokenizer function from nltk.tokenize module over example string 
# returns a list of 3 sentence strings
```

```python
# Applying word tokenizer function
word_tokenize(example_string)
# returns a list of 53 string values - words

# len(word_tokenize(example_string))
```

#### <u>Filtering Stop words:</u> filtering out words that can be ignored

```python
from nltk.corpus import stopwords
```

```python
# Another example string
worf_quote = "Sir, I protest. I am not a merry man!"

#word tokenization
worf_quote_tokens = word_tokenize(worf_quote)
worf_quote_tokens
```

```python
# Initializing stop words set
stop_words = set(stopwords.words("english"))
```

```python
# Initializing an empty list to store words that make it past the filter
filtered_list = []

# Total words in worf_quote
print("Total number of words in worf_quote tokens list:",len(worf_quote_tokens))

# filtering words from the worf_quote by each one and appending to the empty list

for word in worf_quote_tokens:
    if word.casefold() not in stop_words:
        filtered_list.append(word)

# Total words in worf_quote after filtering
print("Total number of words in worf_quote tokens list after filtering the stop words:",len(filtered_list))
print("Filtered worf_quote tokens list:",filtered_list)
```

##### Above filtering process, filters out two words 'I' and 'not' from the original word tokens. Where in general both the words can be significant as they correspond with the content words and context words. 
##### Content words: Informs topics covered in the text or sentiment that author carries about the topic
##### Context words: Informs about the patterns in how authors use context words in order to quantify their writing style.

### <u>Stemming:</u> text processing to reduce words to their root.
##### NLTK has more than 1 stemmer - (Porter, Snowball, ARLSTem), In this course we are using Porter Stemmer


```python
# Import PorterStemmer
from nltk.stem import PorterStemmer

# Initialize PorterStemmer
stemmer = PorterStemmer()
```

```python
# Example string
string_for_stemming = """ The crew of the USS Discovery discovered many discoveries. Discovering is what explorers do."""
```

```python
words = word_tokenize(string_for_stemming)
print(words)
```

```python
# list of the stemmed versions of the words
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

#### Understemming and overstemming are two ways stemming can go wrong:
##### 1. Understemming happens when two related words should be reduced to the same stem but aren’t. This is a false negative.
##### 2. Overstemming happens when two unrelated words are reduced to the same stem even though they shouldn’t be. This is a false positive.

```python
# Working on the example string again with Snowball Stemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.arlstem import ARLSTem


print("Original list of words before stemming:",words)
print("List of words after Porter stemming:",stemmed_words)
# Initialize PorterStemmer
stemmer2 = SnowballStemmer("english")
stemmer3 = ARLSTem()

# list of the stemmed versions of the words
stemmed_words2 = [stemmer2.stem(word) for word in words]

print("List of words after Snowball stemming:",stemmed_words2)

# list of the stemmed versions of the words
stemmed_words3 = [stemmer3.stem(word) for word in words]

print("List of words after ARLSTem stemming:",stemmed_words3)
```

### <u>Lemmatizing:</u>
#### Like stemming, lemmatizing reduces words to their core meaning, but it will give you a complete English word that makes sense on its own instead of just a fragment of a word like 'discoveri'.
#### A lemma is a word that represents a whole group of words, and that group of words is called a lexeme.

For example, if you were to look up the word “blending” in a dictionary, then you’d need to look at the entry for “blend,” but you would find “blending” listed in that entry.

In this example, “blend” is the lemma, and “blending” is part of the lexeme. So when you lemmatize a word, you are reducing it to its lemma.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
# A plural noun
lemmatizer.lemmatize("scarves")
```

```python
# example string
string_for_lemmatizing = "The friends of DeSoto love scarves."

# word tokenization
words = word_tokenize(string_for_lemmatizing)
print("List of words:",words)

#lemmatization
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

print("List of words after lemmatizing:",lemmatized_words)
```

```python
print(lemmatizer.lemmatize("worst"))

# After using lemmatize with a POS tag
print(lemmatizer.lemmatize("worst", pos="a"))
```

#### <u>Chunking:</u> While tokenizing identifies words and sentences, chunking identifies phrases.

Note: A phrase is a word or group of words that works as a single unit to perform a grammatical function. Noun phrases are built around a noun.

Here are some examples:

“A planet”
“A tilting planet”
“A swiftly tilting planet”

```python
# example string 
lotr_quote = "It's a dangerous business, Frodo, going out your door."


# word tokenization
words_in_lotr_quote = word_tokenize(lotr_quote)
print("Tokenized words in the string:",words_in_lotr_quote)

# POS tagging
lotr_pos_tags = nltk.pos_tag(words_in_lotr_quote)
print("POS tagged words: ", lotr_pos_tags)
```

#### In order to chunk, we first need to define a chunk grammar. A chunk grammar is a combination of rules on how sentences should be chunked. It often uses regular expressions, or regexes.

```python
grammar = "NP: {<DT>?<JJ>*<NN>}"

# the above grammer rule is defined as NP chunking which stands for noun phrase chunking, 
# Start with an optional (?) determiner ('DT')
# Can have any number (*) of adjectives (JJ)
# End with a noun (<NN>)
```

```python
chunk_parser = nltk.RegexpParser(grammar)
```

```python
tree = chunk_parser.parse(lotr_pos_tags)
tree.draw()
```

#### Chunking makes use of POS tags to group words and apply chunk tags to those groups. Chunks don’t overlap, so one instance of a word can be in only one chunk at a time.

#### <u>Chinking</u>: Chinking is used together with chunking, but while chunking is used to include a pattern, chinking is used to exclude a pattern.

```python
# similar to chunking we determine grammer to chinking as well
# In chinking REGEXP with more than one line is used - so we use triple quotes

grammar = """ Chunk: {<.*>+} 
}<JJ>{"""
```

```python
chunk_parser = nltk.RegexpParser(grammar)
```

```python
tree = chunk_parser.parse(lotr_pos_tags)
print(tree)
```

```python
tree.draw()
# Output demonstrates that we have excluded the adjective 'dangerous' from the chunk
```

#### <u>Named Entity Recognition (NER):</u>

##### Named entities are noun phrases that refer to specific locations, people, organizations, and so on. With named entity recognition, you can find the named entities in your texts and also determine what kind of named entity they are



```python
# using the above grammer rule, 
tree = nltk.ne_chunk(lotr_pos_tags)
```

```python
tree.draw()
```

```python
# If we do not want to determine what king of Named Entity but want to recognize Named Entities, we use binary=True arugment in the ne_chunk() function
tree = nltk.ne_chunk(lotr_pos_tags, binary=True)
tree.draw()
# the output shows the Named Entity, which is 'Frodo' which is the only one mentioned in the sentence but is not named as 'PERSON'
```

```python
# Example on a bigger sentence. 
quote = """
Men like Schiaparelli watched the red planet—it is odd, by-the-bye, that for countless centuries Mars has been the star of war—but failed to interpret the fluctuating appearances of the markings they mapped so well. All that time the Martians must have been getting ready. 
During the opposition of 1894 a great light was seen on the illuminated part of the disk, first at the Lick Observatory, then by Perrotin of Nice, and then by other observers. English readers heard of it first in the issue of Nature dated August 2."""
```

```python
# A function to extract named entities

def extract_ne(quote):
    words = word_tokenize(quote)
    tags = nltk.pos_tag(words)
    tree = nltk.ne_chunk(tags, binary=True)
    return set(
        " ".join(i[0] for i in t)
        for t in tree
        if hasattr(t,"label") and t.label() == "NE"
    )


extract_ne(quote)
```
