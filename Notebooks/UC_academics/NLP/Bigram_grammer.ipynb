```python
import spacy
# Load the English language model
nlp = spacy.load("en_core_web_sm")
```

```python

# Original sentence
original_sentence = "The quick brown fox jumps over the lazy dog and threatens it for its life."

# Process the original sentence
doc = nlp(original_sentence)
original_tags = [(token.text, token.pos_) for token in doc]
```

```python
# Introduce spelling errors
misspelled_sentence = "The quik brown fox jmps ovr the lazi dawg and thretens it for its lyf."


# Process the misspelled sentence
doc = nlp(misspelled_sentence)
misspelled_tags = [(token.text, token.pos_) for token in doc]
```

```python

# Print the results
print("Original Sentence:")
print(original_sentence)
print(original_tags)

print("\nMisspelled Sentence:")
print(misspelled_sentence)
print(misspelled_tags)
```
