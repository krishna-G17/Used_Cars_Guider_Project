```python
from transformers import AutoTokenizer
from collections import Counter
import collections
import nltk
from nltk.corpus import stopwords
```

```python

# Load the text file
file_path = 'Bigram_project_text.txt'
with open(file_path, 'r') as file:
    text = file.read()

# Initialize the tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Tokenize the text by word
tokens = tokenizer.tokenize(text)

# Create bigrams
bigrams = [(tokens[i], tokens[i+1]) for i in range(len(tokens) - 1)]

# Count bigrams
bigram_counts = Counter(bigrams)

# Get the 10 most common bigrams
most_common_bigrams = bigram_counts.most_common(10)

# Print the results
for bigram, count in most_common_bigrams:
    print(f'Bigram: {bigram}, Count: {count}')
```

```python
import collections

# Read the text file
with open('Bigram_project_text.txt', 'r') as file:
    text = file.read()

# Split the text into words and remove empty strings
words = [word for word in text.split() if word]

# Create two lists: words_shifted and next_words
words_shifted = words[:-1]  # All words except the last one
next_words = words[1:]      # All words except the first one

# Zip the two lists together to create bigrams
bigrams = list(zip(words_shifted, next_words))

# Count the frequency of each bigram
bigram_counts = collections.Counter(bigrams)

# Print the 10 most common bigrams
print("10 Most Common Bigrams:")
for bigram, count in bigram_counts.most_common(10):
    print(f"{bigram}: {count}")
```

```python



# Read the text file
file_path = 'Bigram_project_text.txt'
with open(file_path, 'r') as file:
    text = file.read()

# Split the text into words and remove empty strings
words = [word for word in text.split() if word]

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]

# Create two lists: filtered_words_shifted and next_filtered_words
filtered_words_shifted = filtered_words[:-1]  # All words except the last one
next_filtered_words = filtered_words[1:]      # All words except the first one

# Zip the two lists together to create bigrams
bigrams = list(zip(filtered_words_shifted, next_filtered_words))

# Count the frequency of each bigram
bigram_counts = collections.Counter(bigrams)

# Print the 10 most common bigrams
print("10 Most Common Bigrams:")
for bigram, count in bigram_counts.most_common(10):
    print(f"{bigram}: {count}")
```
