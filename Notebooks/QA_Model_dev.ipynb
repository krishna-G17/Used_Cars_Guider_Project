```python
import pandas as pd 
import torch
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline
```

```python
QA_input = [{'question':'Why is conversion important?',
'context':'The option to convert models between FARM and transfomers gives freedom to the user between frameworks'},
{'question':'How many programming languages does BLOOM support?',
'context':'BLOOM is a state-of-the-art language model that is trained on a large dataset of text. BLOOM has 175 billion parameters and can generate in 46 languages.'}]
```

```python
model_name = "deepset/roberta-base-squad2"
```

```python
# Method - 1
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
inputs0 = tokenizer(QA_input[0]['question'], QA_input[0]['context'], return_tensors="pt")
output0 = model(**inputs0)
```

```python
output0
```

```python
answer_start_idx = torch.argmax(output0.start_logits)
answer_end_idx = torch.argmax(output0.end_logits)

answer_tokens = inputs0.input_ids[0, answer_start_idx:answer_end_idx+1] # [answer_start_idx:answer_end_idx+1]
answer = tokenizer.decode(answer_tokens)
print("Ques: {}\nanswer: {}".format(QA_input[0]['question'], answer))
```
